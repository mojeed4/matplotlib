```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

X = np.linspace(-14, 14, 1024)
Y = np.sinc(X)

ax = plt.axes()
ax.xaxis.set_major_locator(ticker.MultipleLocator(5))
ax.xaxis.set_minor_locator(ticker.MultipleLocator(1))

plt.grid(True, which='both')
plt.plot(X, Y)
plt.show()
```

```python
name_list = ('Mike', 'Brad', 'Stan', 'Jon', 'Matt')
value_list = np.random.randint(0, 99, size = len(name_list))
pos_list = np.arange(len(name_list))

ax = plt.axes()
ax.xaxis.set_major_locator(ticker.FixedLocator((pos_list)))
ax.xaxis.set_major_formatter(ticker.FixedFormatter((name_list)))

plt.bar(pos_list, value_list, color = '.75', align = 'center')
plt.show()
```

```python
plt.bar(pos_list, value_list, color = '.5', align = 'center')
plt.xticks(pos_list, name_list)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
sns.swarmplot(x = "species", y = "petal_length", data = iris)
plt.show()
```

```python
data = np.random.normal(size = (20, 6)) + np.arange(6)/2
sns.set_style("whitegrid")
sns.boxplot(data = data)
plt.title("whitegrid")
plt.show()

sns.set_style("darkgrid")
sns.boxplot(data = data)
plt.title("darkgrid")
plt.show()

sns.set_style("white")
sns.boxplot(data = data)
plt.title("white")
plt.show()

sns.set_style("dark")
sns.boxplot(data = data)
plt.title("dark")
plt.show()

sns.set_style("ticks")
sns.boxplot(data = data)
plt.title("ticks")
plt.show()
```

```python
titanic = sns.load_dataset("titanic")
g = sns.factorplot("class", "survived", "sex", data = titanic, kind = "bar", palette = "muted", legend = True)
plt.show()
```

```python
from matplotlib.colors import ListedColormap

N = 500

current_palette = sns.color_palette("muted", n_colors = 5)
cmap = ListedColormap(sns.color_palette(current_palette).as_hex())

data1 = np.random.randn(N)
data2 = np.random.randn(N)

colors = np.random.randint(0, 5, N)
plt.scatter(data1, data2, c = colors, cmap=cmap)
plt.colorbar()
plt.show()
```
